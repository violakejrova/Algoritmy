```python
seznam = [6,4,1,666,42,-4,15]

serazeno = False

while serazeno == False:
  for n in range(len(seznam)):

    for i in range (0,(len(seznam))-1):
      if seznam [i] > seznam [i+1]:
        seznam [i], seznam [i+1] = seznam[i+1], seznam [i]
        i += 1
        serazeno = False
      else:
        i+=1
        serazeno = False

  serazeno = True

print(seznam)
```
